# Full 'Exception Handling Block'

```python
while True:
    no_one = int(input("The first number please : "))
    no_two = int(input("The second number please : "))
    try:
        division = no_one / no_two  # normal part of the program
    except ZeroDivisionError:
        print("You can't divide by zero! Try again.")  # executes when division by zero
    else:
        print("The result of the division is : ", division)  # executes if there is no exception
    finally:
        print("Thanks for using our mini divison calculator! Come again!") # executes always
        break  # exits the while loop
```

```python
fruits = ["banana", "mango", "pear", "apple", "kiwi", "grape"]
counter = 3

while counter > 0:
    try:
        print(f"You have {counter} right.")
        index = int(input("ndex numaras giriniz: "))
        print("Your favorite fruit is :", fruits[index])
              
    except IndexError:
        counter -= 1
        if counter != 0:
            print(f"Index error raised. You have {counter} left. Try again!")
        else:
            print(f"Index error raised. Sorry you have no right left.")
              
    except ValueError:
        counter -= 1
        if counter != 0:
            print(f"Value error raised. You have {counter} left. Try again!")
        else:
            print(f"Value error raised. Sorry you have no right left.")
              
    else: #exeption ykselmezse
        print("Congrats! You have entered valid input")
        break
        
    finally:
        print ("Our fruits are always fresh!")
```

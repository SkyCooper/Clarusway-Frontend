```python
# LERLEME YNNE DORU START STOPTAN NCE OLACAK, negatif pozitif fark etmez, kar olabilir;
# [1:-5]  [3:-2] [-7:3] [-5:-2] hepsi olur.

# LERLEM YN POZTF STEPTE SOLDAN SAA / NEGATF STEPTE SADAN SOLA DORU

reef = ['swordfish', 'shark', 'whale', 'jellyfish', 'lobster', 'squid', 'octopus']
#           0             1      2          3           4         5         6
#           -7           -6     -5          -4         -3         -2        -1
print(reef[-3:])
print(reef[4:])

print(reef[:-3])
print(reef[:4])

print(reef[::-1])  # we have produced the reverse of the list TERSTEN YAZDI
print(reef[::-2])
```

```python
# and ile ilgili rnekler


print(2 and 3)                   # true and true = true
print(1 and 0)                   # true and false = false
print(1 and "I am doing good!")  # true and true = true
print([] and "Hello World!")     # false and true = false
print([] and 0)                  # false and false = false
```

```python
# or ile ilgili rnekler
print(2 or 3)               # true or true = true
print(None or 1)            # false or true = true
print(0 or {})              # false or false = false
print([] or "Hello World!") # false or true = true
```

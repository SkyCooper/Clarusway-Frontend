```python
max(numbers, key = numbers.count) 

# max foksiyonunun default değerini key parametresi ile değiştiriyoruz.
# key parametresine atanan foksiyonun max değerini getir. --> DEMEKTİR.
# numbers'ın bütün elemanlarını tek tek sayar, çıkan sonuçlardan max olan count değerini getir.
# 3 sayısı 4 kere ile en çok count olacağı için sonuç olarak count(3)--->4 gibi görür.
# sonuç olarak en çok tekrar edeni buluruz.
```

```python
numbers.count(max(numbers, key = numbers.count)) # aslında içine 3 yazmamız lazım
# ama içine bu şekilde yazabiliriz. yani en çok tekrar eden sayı 4 kere tekrar etmiş.
```

```python
# me-decoders ortak çözümü-1
numbers = [1, 3, 7, 4, 3, 0, 3, 6, 6, 3]
result = max(numbers, key = numbers.count)
print("the most freaquant number", result, end=', ')
print("it was repeated", numbers.count(result), "times.")
```

```python
# me-decoders ortak çözümü-2
numbers = [1, 3, 7, 4, 3, 0, 3, 6, 6, 3]
frequant_num = max(numbers, key = numbers.count)
freaquency = numbers.count(frequant_num)
print(f"the most freaquant number is {frequant_num} and it was repeated {freaquency} times.")
```

```python
# key parametresi başka bir örnek

şehirler = ["istanbul", "ağrı", "ankara", "van"]
```

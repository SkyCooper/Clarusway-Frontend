```python
my_tup = ("Romania", "Poland", "Estonia")
(ro, po, es) = my_tup # elaman saylar ayn olmal, tek tek atama yapar python
print(ro + ", " + po + ", " + es) #returns 'Romania, Poland, Estonia'
```

```python
my_tup = ("Romania", "Poland", "Estonia")
ro = my_tup[0] # tek tek indexe gre atama yaplabilir
po = my_tup[1]
es = my_tup[2]
print(ro + ", " + po + ", " + es)
```

```python
# max ve min kullanm
my_tup = ("Romania", "Poland", "Estonia", "Bulgaria", "Slovakia", "Slovenia", "Hungary") 
last = max(my_tup)   # en byk deeri sralama yapmadan verir
# last = sorted(my_tup)[-1] sralama yapar ve en son indexi arr

first = min(my_tup)  # en kk deeri sralama yapmadan verir
# first = sorted(my_tup)[0] sralama yapar ve ilk indexi arr
print(last, first)
```

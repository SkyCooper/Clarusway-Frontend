# Lambda within User-Defined Functions

```python
def modular_function(n):
    return lambda x: x ** n
    
power_of_2 = modular_function(2)  # first sub-function derived from def
power_of_3 = modular_function(3)  # second sub-function derived from def
power_of_4 = modular_function(4)  # third sub-function derived from def

print(power_of_2(2))  # 2 to the power of 2
print(power_of_3(2))  # 2 to the power of 3
print(power_of_4(2))  # 2 to the power of 4
```

```python
# most frequant sorusu;
num = [1, 1, 1, 4, 5, 5]
max(num, key = num.count) # max numaray getir ama en ok tekrar says byk olan getir. demek 
                          # 1 says 3 kere tekrarla en ok tekrar eden olduu inin cevap 1
```

```python
num.count(max(num, key = num.count))  
```

```python
def equal(a, b, c):
    num = [a, b, c]
    if num.count(max(num, key = num.count)) > 1:
        return  num.count(max(num, key = num.count))
    else:
        return 0
    
equal(1, 4, 4)
```

```python
def equal_(*a):
    num = list(a)
    if num.count(max(num, key = num.count)) > 1:
        return  num.count(max(num, key = num.count))
    else:
        return 0
    
equal_(1, 4, 4, 7, 9, 5, 5, 2, 5, 3, 5)
```
